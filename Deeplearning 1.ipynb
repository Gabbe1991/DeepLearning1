```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


model = ResNet50(weights='imagenet')


img_paths = ['/content/Blomma.jpg', '/content/Glas.jpg', '/content/Mugg.jpg' , '/content/Avocado.jpg']


for img_path in img_paths:

    img = mpimg.imread(img_path)
    plt.imshow(img)
    plt.axis('off')  # Turn off axes
    plt.show()

    # Laddar
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)


    preds = model.predict(x)


    decoded_preds = decode_predictions(preds, top=3)[0]


    print("Predicted classes for", img_path)
    for i, (imagenet_id, label, score) in enumerate(decoded_preds):
        print(f"{i+1}. {label} ({score:.2f})")
    print()




    # 2. Answer the question: Why is "Pretrained models" a very powerful concept?
    #Pretrained models are like having a super-smart assistant who's already learned a ton of stuff.
    # They come pre-equipped with knowledge gained from analyzing vast amounts of data, which saves loads of time and effort. Take, for instance, a language model that's been trained on millions of texts; it can help with tasks like writing,
    # summarizing, and even translating. With pretrained models, you're essentially leveraging the collective wisdom of the internet, making your own projects smarter and more efficient.


```
